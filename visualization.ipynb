```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
import torch
torch.load('models/standard_107/standard_107_acc_history.npy')
```

```python
acc_history = np.load('models/standard_107/standard_107_acc_history.npy') # This is univSample_107
for i in range(acc_history.shape[0]):
    plt.plot(np.arange(1, acc_history.shape[1]+1), acc_history[i,:])
plt.legend(['Train Accuracy', 'Test Accuracy', 'Test Accuracy Vote'])
plt.show()

print('Epoch: %d, Best test vote acc: %f' % (np.argmax(acc_history[2,:])+1, np.max(acc_history[2,:])))
print('Epoch: %d, Best test acc: %f' % (np.argmax(acc_history[1,:])+1, np.max(acc_history[1,:])))
```
