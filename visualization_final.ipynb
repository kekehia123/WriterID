```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def plot_acc_loss(model_name):
    acc_history = np.load(os.path.join('models', model_name, model_name+'_acc_history.npy'))
    for i in range(3):
        plt.plot(np.arange(1, acc_history.shape[1]+1), acc_history[i,:])
    plt.legend(['Training Accuracy', 'Validation Accuracy', 'Validation Accuracy Vote'])
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.show()

    for i in range(3,5):
        plt.plot(np.arange(1, acc_history.shape[1]+1), acc_history[i,:])
    plt.legend(['Training loss', 'Validation loss'])
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.show()

    print('Epoch: %d, Best validation vote acc: %f' % (np.argmax(acc_history[2,:])+1, np.max(acc_history[2,:])))
    print('Epoch: %d, Best validation acc: %f' % (np.argmax(acc_history[1,:])+1, np.max(acc_history[1,:])))
```
